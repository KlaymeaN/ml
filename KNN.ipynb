# K Nearest Neighbours
## ASSUMPTION of this Algorithm: data points that are similar, have similar labels

In this notebook, we are going to implement the Nearest neighbours alrogithm,
in the simplest way, using minimum imports, so I am importing numpy for calculations,
pandas to read the data and matplotlib to plot the datapoints for better intuition.
a simple explanation of the algorithm can be stated as follows:



### Algorithm:
* Trainig: Store all the training data (remember it).
* When we are presented a test point, we compute the distance of its features to all the same features of the trianing data, and sum them all up for each training point to have a single distance value to the test point,
* Then we sort the distances in ascesding order, meaning the point with the lowest distances will be at the top of our list.
* Then we choose an arbitraty number of them from the top, namely K ones.
* Those are our nearest neighbours.
* Then based on their labels, we take the majority vote of what our test point is likely to be

### How to comupte the distance? 
There are several methods, but here I got over two simple ones:
* Manhattan distance: Just subtract them
$$
\Huge \boldsymbol{
\sum_{i=1}^{n} \left| x_{\text{train},i} - x_{\text{test},i} \right|
}
$$

* Euclidian distance: Everyone is familiar with it
$$ 
\Huge d = \boldsymbol{
\sqrt{ \sum_{i=1}^{n} \left(x_{\text{train}, i} - x_{\text{test}, i} \right)^2 }
}
$$

I have implemented both of them here, and you can choose which one you want your KNN to be using

### So we have 3 classes

```python
df = df.sample(frac=1, random_state=13).reset_index(drop=True)

df_len = (len(df))
df_train = df[:int(0.85*df_len)]
df_test  = df[int(0.85*df_len):]

X_train = df_train[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y_train = df_train['Species']

X_test = df_test[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y_test = df_test['Species']

print(len(df_train), len(df_test))
```

```python
def dist_man(x,y): #Manhattan distance
    return sum(abs(x-y))

def dist_euc(x,y): #Euclidian distance
    d = sum(np.sqrt(abs((y-x)**2)))
    return d
```

```python
d = []
x_t = X_test.iloc[5]
for i in range(len(X_train)):
    dis = (dist_euc(X_train.iloc[i] , x_t))
    d.append((dis, y_train.iloc[i]))
    
#d 
```

```python
d.sort(reverse=False)
print(len(d))
#d
```

```python
k = 10
neighbours = d[:k]
```

```python
c_0 = 0
c_1 = 0
c_2 = 0

for i in range(k):
    if neighbours[i][1] == 'Iris-setosa':
        c_0 +=1
    elif neighbours[i][1] == 'Iris-versicolor':
        c_1 += 1 
    elif neighbours[i][1] == 'Iris-virginica':
        c_2 +=1

if c_2 == max(c_0, c_1 , c_2):
    print('Virginica')
elif c_1 == max(c_0, c_1 , c_2):
    print('Versicolor')
else:
    print('Virginia')
```

## Now lets write a class that we can make it more general

```python
class KNN(object):
    def __init__(self):
        pass
    def train(self, X, y):
        self.X_train = X
        self.y_train = y
        classes = []
        self.classes = classes
        for i in y.unique():
            classes.append(i)
        clas = np.zeros(len(classes))
        self.clas = clas
        

    def predict(self, xt, neighbours, dist_method, verbose=False):
        d = []
        for i in range(len(self.X_train)):
            #d = []
            dis = dist_method(self.X_train.iloc[i], xt)
            d.append((dis, self.y_train.iloc[i]))
        d.sort(reverse=False)
        neigh = d[:neighbours]

        for i in range(len(neigh)):
            for j in range(len(self.classes)):
                if neigh[i][1] == self.classes[j]:
                    self.clas[j] +=1
        if verbose:
            print(self.clas, self.classes)
        predicted_index = np.argmax(self.clas)
        predicted_class = self.classes[predicted_index]
        return predicted_class

    def get_neighbours(self, xt, neighbours, dist_method):
        # a function to make 3D plotting easier
        d = []
        for i in range(len(self.X_train)):
            dis = dist_method(self.X_train.iloc[i], xt)
            #d.append((dis, self.y_train.iloc[i]))
            d.append((dis, i))
        d.sort(reverse=False)
        neigh = []
        for i in range(neighbours):
            _, idx = d[i]
            neigh.append(idx)
        return neigh
        
        
            
        
        
```

### Lets use our class!

```python

knn = KNN()
knn.train(X_train, y_train)
t= 12
x_t = X_test.iloc[t]
true_label = y_test.iloc[t]
knn.predict(x_t, 5, dist_euc, verbose=True)

print(true_label)
```

 Here I chose 3 features so it can be easily plottable in a 3D space ( each features can be interpreted as a 
dimension, and you can change one with the fourth feature to see that one too!

```python
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm']

t= 7 #test num
n_n = 5 #num of neighbours
x_t = X_test.iloc[t]
true_label = y_test.iloc[t]

X_3d = X_train[features]
pred = knn.predict(x_t, n_n, dist_euc)
neighbor_indices = knn.get_neighbours(x_t, n_n, dist_euc)

#%matplotlib notebook
fig= plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

for cls in y_train.unique():
    ix = y_train[y_train == cls].index
    ax.scatter(X_3d.loc[ix, features[0]],
               X_3d.loc[ix, features[1]],
               X_3d.loc[ix, features[2]],
               label=cls)

ax.scatter(x_t[features[0]], x_t[features[1]], x_t[features[2]], c='black', marker='X', 
           s = 100, label='Test Point')


for idx in neighbor_indices:
    n = X_train.loc[idx][features]
    ax.scatter(n[features[0]], n[features[1]], n[features[2]],
               edgecolor='red', facecolor='none', s=100, marker='o', linewidths=2)

ax.set_xlabel(features[0])
ax.set_ylabel(features[1])
ax.set_zlabel(features[2])
ax.legend()
plt.title(f'KNN Prediction ({n_n} neighbours)\nTrue: {true_label}, Predicted: {pred}')


plt.show()
```
